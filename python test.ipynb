Write a program that accepts space separated data. For the entered data, extract all the prime numbers.

```python
n=list(map(int,input().strip().split(" ")))
for i in n:
  if i==0 or i==1 or i==2:
    print(i,"is a prime")
  flag=0
  for j in range(2,i):
    if i%j==0:
      flag=1
      break
  if flag==1:
    print(i," is not prime")
  elif flag==0:
    print(i,"is prime")
```

With the help of classes and objects create a program to add, subtract, multiply and divide the data.

```python
class calculator:
    def add(self,a,b):
        print(a+b)
    def mul(self,a,b):
        print(a*b)
    def div(self,a,b):
        print(a/b)
    def sub(self,a,b):
        print(a-b)


obj = calculator()
obj.add(3,7)
obj.mul(45,67)
obj.div(45,98)
obj.sub(54,89)
```
