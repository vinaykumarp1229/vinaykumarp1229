# Seaborn Continuation

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# loading the tips dataset
data1 = sns.load_dataset("tips")
```

```python
data1.head(20)
```

```python
# To know the available list of dataset present in sns - seaborn
sns.get_dataset_names()
```

```python
data1.info()
```

# count plot

```python
# It is used to show the count of observation based upon each category using bars.
```

```python
data1.value_counts('sex')
```

```python
sns.countplot('sex', data = data1)
plt.show()
```

```python
data1.value_counts('smoker')
```

```python
sns.countplot(x = 'smoker', data = data1)  # bydefault it is x = 'smoker'
plt.show()
```

```python
# lets change the axis
sns.countplot(y = 'smoker', data = data1)
plt.show()
```

# Box Plot

```python
# We plot box plot to see the outliers in our 
# box plot is applicable for numerical columns but we can take the reference of categorical column
# for the bifurcation of numerical column.
# https://prnt.sc/MpaOKqxMclS0
```

```python
sns.boxplot('total_bill', data = data1)
plt.show()
```

```python
sns.boxplot(y ='total_bill', data = data1)
plt.show()
```

```python
# lets customize it
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = data1)
plt.show()
```

```python
data1.value_counts('day')
```

```python
# it is dividing the total_bill column into two categories
# one for male and one for female
# and then individually plotting the box_plot.
sns.boxplot(x = 'sex', y = 'total_bill', data = data1)
plt.show()
```

```python
sns.boxplot(y = 'sex', x = 'total_bill', data = data1)
plt.show()
```

```python
# lets make it more difficult.
```

```python
sns.boxplot(x = 'sex', y = 'total_bill', hue = 'smoker',data = data1)
plt.show()

# hue is a target variable
```

```python
sns.boxplot(x = 'day', y = 'total_bill', hue ='smoker', data = data1)
plt.show()
```
